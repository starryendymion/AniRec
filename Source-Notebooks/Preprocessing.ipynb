```python
import pandas as pd
```

```python
anime=pd.read_csv("animes.csv").dropna().drop_duplicates()
```

```python
anime.dtypes
```

```python
anime.sample(5)
```

```python
filtered_anime= anime[(anime['Rating Score']!="Unknown") & (anime['Name']!="Unknown") & (anime['Anime-PlanetID']!="Unknown")]
```

```python
filtered_anime
```

```python
filtered_anime= filtered_anime[(filtered_anime['Rating Score'].astype(float) > 3.5) & (filtered_anime['Type'] == 'TV')]
```

```python
filtered_anime = filtered_anime.rename(columns={'Anime-PlanetID': 'anime_id'})
```

```python
filtered_anime
```

```python
ratings=pd.read_csv("ratings.csv").dropna().drop_duplicates()
```

```python
ratings.dtypes
```

```python
ratings.sample(2)
```

```python
filtered_ratings = ratings[ratings['anime_id'].isin(filtered_anime['anime_id'])]
```

```python
filtered_ratings
```

```python
filtered_ratings['user_id'] = filtered_ratings['user_id'].astype(str)
```

```python
filtered_ratings = filtered_ratings[(filtered_ratings['rating']==5.0)]
```

```python
filtered_ratings
```

```python
filtered_ratings = pd.merge(filtered_ratings, filtered_anime[['anime_id', 'Name']], on='anime_id', how='left')
```

```python
filtered_ratings
```

```python
filtered_anime.to_csv("processed_animes.csv")
```

```python
filtered_ratings.to_csv("filtered_ratings.csv")
```
